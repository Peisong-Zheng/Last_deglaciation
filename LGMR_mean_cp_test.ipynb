# Load the LGMR mean

```python
import xarray as xr
# open the 'LGMR_GMST_ens.nc' file
ds_sat = xr.open_dataset('data/LGMR_SAT_climo.nc')
ds_sat
```

# Crop the data for age>11

```python

# ds_sat = ds_sat.where(ds_sat.age>=11000, drop=True)

ds_sat = ds_sat.where(ds_sat.age>=11000, drop=True)
```

```python
ds_sat.info()
```

```python
from toolbox import change_point_functions as cpf

import numpy as np
import matplotlib.pyplot as plt


# Initialize an array to hold the change point ages
change_points = np.zeros((len(ds_sat['lat']), len(ds_sat['lon'])))

# Iterate over latitude and longitude
for i, lat in enumerate(ds_sat['lat'].values):
    for j, lon in enumerate(ds_sat['lon'].values):
        # Extract the time series for this lat-lon point
        time_series = ds_sat['sat'][:, i, j].values

        # Apply the find_cp function
        # Assuming 'age' is a coordinate in ds_sat with the same size as the 'sat' variable
        output = cpf.find_cp(time_series, ds_sat['age'].values)

        # Store the change point age
        change_points[i, j] = output['cp_age']

# Plotting the heatmap of change points
plt.figure(figsize=(10, 8))
plt.imshow(change_points, cmap='viridis', origin='lower')
plt.colorbar(label='Change Point Age (yr BP)')
plt.xlabel('Longitude Index')
plt.ylabel('Latitude Index')
plt.title('Heatmap of Change Points in SAT Data')
plt.show()
```
