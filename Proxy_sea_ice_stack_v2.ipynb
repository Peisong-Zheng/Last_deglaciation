Ref: New data from M. Osman

```python
path='D:/VScode/Last_deglaciation/data/LGMR_SIC_May2023.nc'

# read the nc file from the path
import xarray as xr
import numpy as np

ds_seaice = xr.open_dataset(path)
ds_seaice = ds_seaice.where(ds_seaice.age>=11000, drop=True)

# Calculate the weight based on latitude
weight = np.cos(np.deg2rad(ds_seaice['lat']))
print('shape of the weight:', weight.shape)

weight = xr.where(weight < 0, 0, weight)

ds_seaice['weight'] = weight
ds_seaice
```

```python
import xarray as xr
import numpy as np
from scipy.stats import zscore

sector_ranges = {
    'Scotia Sea': (-60 + 360, -30 + 360),  # Adjusting for a 0–360 longitude system
    'Central Atlantic': (-30 + 360, 0 + 360),
    'East Atlantic': (0, 30),
    'Indian': (30, 60),
    'Pacific': (150, 255)  # 255 is 360 - 105
}

# Select the SIC data of given latitude band
selected_data = ds_seaice['sic'].isel(dims=1).sel(lat=slice(-70, -50))
weights = ds_seaice['weight'].sel(lat=slice(-70, -50))

# Expand the weights to match the dimensions of the data
weights_expanded = weights.expand_dims(dim={'lon': selected_data.lon, 'age': selected_data.age}, axis=[1, 2])

# Calculate the weighted average for each sector
weighted_avg_sic = {}

for sector, (lon_start, lon_end) in sector_ranges.items():
    lon_slice = slice(lon_start, lon_end)
    data_group = selected_data.sel(lon=lon_slice)
    weights_group = weights_expanded.sel(lon=lon_slice)

    # Remove the data and weights where data is all zero
    data_group = data_group.where(data_group > 0, drop=True)
    weights_group = weights_group.where(data_group > 0, drop=True)

    # Calculate the weighted average
    weighted_avg = (data_group * weights_group).sum(dim=['lat', 'lon']) / weights_group.sum(dim=['lat', 'lon'])
    weighted_avg_sic[sector] = weighted_avg.values


# Convert the dictionary values to a list of arrays
sic_curves = list(weighted_avg_sic.values())


def min_max_scaling(array):
    min_val = np.min(array)
    max_val = np.max(array)
    scaled_array = (array - min_val) / (max_val - min_val)
    return scaled_array

# Apply Min-Max scaling to each array in sic_curves
sic_curves_normalized = [min_max_scaling(curve) for curve in sic_curves if curve.size > 0]

# Then you can calculate the mean and standard deviation of these normalized curves
mean_sic_curve = np.mean(sic_curves_normalized, axis=0)
std_sic_curve = np.std(sic_curves_normalized, axis=0)
#################################################################################
# # Normalize each array in sic_curves using z-score
# sic_curves_normalized = [zscore(curve) for curve in sic_curves if curve.size > 0]  # Ensures that the curve has elements

# # Check that all arrays are of the same length
# if all(len(curve) == len(sic_curves_normalized[0]) for curve in sic_curves_normalized):
#     # Calculate the mean curve and its standard deviation (as error range) from the normalized curves
#     mean_sic_curve = np.mean(sic_curves_normalized, axis=0)
#     std_sic_curve = np.std(sic_curves_normalized, axis=0)
# else:
#     raise ValueError("The SIC curves are not of the same length.")
###################################################################################
# # Check that all arrays are of the same length
# if all(len(curve) == len(sic_curves[0]) for curve in sic_curves):
#     # Calculate the mean curve and its standard deviation (as error range)
#     # sic_curves=zscore(sic_curves)
#     mean_sic_curve = np.mean(sic_curves, axis=0)
#     std_sic_curve = np.std(sic_curves, axis=0)
# else:
#     raise ValueError("The SIC curves are not of the same length.")


```
