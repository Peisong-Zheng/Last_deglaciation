# Read LGMR SIC data
Ref: New data from M. Osman

```python
path='D:/VScode/Last_deglaciation/data/LGMR_SIC_May2023.nc'

# read the nc file from the path
import xarray as xr
import numpy as np

ds_seaice = xr.open_dataset(path)
ds_seaice = ds_seaice.where(ds_seaice.age>=11000, drop=True)

# Calculate the weight based on latitude
weight = np.cos(np.deg2rad(ds_seaice['lat']))
print('shape of the weight:', weight.shape)

weight = xr.where(weight < 0, 0, weight)

ds_seaice['weight'] = weight
ds_seaice
```

# Load the LGMR sea ice

```python
import xarray as xr
import numpy as np

sector_ranges = {
    'Scotia Sea': (-60 + 360, -30 + 360),  # Adjusting for a 0â€“360 longitude system
    'Central Atlantic': (-30 + 360, 0 + 360),
    'East Atlantic': (0, 30),
    'Indian': (30, 60),
    'Pacific': (150, 255)  # 255 is 360 - 105
}

# Select the SIC data of given latitude band
selected_data = ds_seaice['sic'].isel(dims=1).sel(lat=slice(-70, -50))
weights = ds_seaice['weight'].sel(lat=slice(-70, -50))

# Expand the weights to match the dimensions of the data
weights_expanded = weights.expand_dims(dim={'lon': selected_data.lon, 'age': selected_data.age}, axis=[1, 2])

# Calculate the weighted average for each sector
weighted_avg_sic = {}

for sector, (lon_start, lon_end) in sector_ranges.items():
    lon_slice = slice(lon_start, lon_end)
    data_group = selected_data.sel(lon=lon_slice)
    weights_group = weights_expanded.sel(lon=lon_slice)

    # Remove the data and weights where data is all zero
    data_group = data_group.where(data_group > 0, drop=True)
    weights_group = weights_group.where(data_group > 0, drop=True)

    # Calculate the weighted average
    weighted_avg = (data_group * weights_group).sum(dim=['lat', 'lon']) / weights_group.sum(dim=['lat', 'lon'])
    weighted_avg_sic[sector] = weighted_avg.values

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cartopy.crs as ccrs

# Defining the sectors and their corresponding standard deviation columns
sectors = ['Scotia Sea', 'Central Atlantic', 'East Atlantic', 'Indian', 'Pacific']
std_dev_columns = ['St. dev.', 'St. dev..1', 'St. dev..2', 'St. dev..3', 'St. dev..4']
colors = ['b', 'g', 'r', 'c', 'm']  # Blue, Green, Red, Cyan, Magenta

sic_colors = {
    'Scotia Sea': 'b',
    'Central Atlantic': 'g',
    'East Atlantic': 'r',
    'West Indian': 'c',
    'Pacific': 'm'
}

# Creating subplots with shared axes
fig, axs = plt.subplots(len(sectors), 1, figsize=(4, 14), dpi=600, sharex=True, sharey=True)

# Remove vertical gap between subplots and set boundaries invisible
fig.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in fig.axes[:-1]], visible=False)
plt.setp([a.spines['top'].set_visible(False) for a in fig.axes[1:]])
plt.setp([a.spines['bottom'].set_visible(False) for a in fig.axes[:-1]])

for i, ax in enumerate(axs):
    sector = sectors[i]
    mean_column = f'{sector} Mean'
    std_dev_column = std_dev_columns[i]

    upper_bound = df[mean_column] + df[std_dev_column]
    lower_bound = df[mean_column] - df[std_dev_column]

    # Plot the mean sea ice concentration in each subplot
    ax.plot(df['Age (ka)']*1000, df[mean_column], color=colors[i])
    if i==0:
        ax.plot(df_pc['Age (ka)']*1000, df_pc['PC 1'], color='orange', marker='^', label='Proxy data PC 1')
        # add legend to the first subplot
        ax.legend(loc='lower left')

    # Fill the range between the upper and lower error range
    ax.fill_between(df['Age (ka)']*1000, lower_bound, upper_bound, color=colors[i], alpha=0.1)
    # Set limits and labels for each subplot
    ax.set_xlim(11000, 30000)
    ax.set_ylabel(f'{sector} sea ice',color=colors[i])
    # add xlabel to the last subplot
    if i == len(sectors)-1:
        ax.set_xlabel('Age (yr BP)')

    ax2 = ax.twinx()
    # Plot the weighted average curve from the simulated data
    ax2.plot(ds_seaice['age'], weighted_avg_sic[sector], color='k', linestyle='-', label=f'Simulated {sector}')
    ax2.set_ylabel(f'{sector} SIC (fraction)')
    ax2.spines['top'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)

# add a vertical line at 21000
for ax in axs:
    # invert the x-axis
    ax.invert_xaxis()
    ax.axvline(x=21000, color='k', linestyle='--', alpha=0.5)


ax_robinson = fig.add_axes([-0.62, 0.58, 0.5, 0.5], projection=ccrs.Robinson())
# Adjusted function to add colored bands according to proxy data sectors
def add_colored_bands(ax):
    extent = [-180, 180, -70, -40]  # Set the extent to cover the relevant latitudes
    ax.set_extent(extent, crs=ccrs.PlateCarree())

    # Define the longitudinal extents for each sector
    sector_extents = {
        'Scotia Sea': (-60, -30),
        'Central Atlantic': (-30, 0),
        'East Atlantic': (0, 30),
        'West Indian': (30, 60),
        'Pacific': (150, -105 + 360)  # Adjusting for the 0-360 longitude system
    }

    # Draw filled polygons for each sector
    for sector, (start_lon, end_lon) in sector_extents.items():
        width = end_lon - start_lon
        ax.add_patch(patches.Rectangle(
            (start_lon, -70), width, 30, 
            transform=ccrs.PlateCarree(),
            color=sic_colors[sector],
            alpha=1  # Adjust transparency as needed
        ))

# Call the function to add colored bands to the map
add_colored_bands(ax_robinson)
ax_robinson.coastlines()
ax_robinson.set_global()



# Periods for the vertical grey bars
HS1 = np.array([17480, 14692]) - 50  # convert to b1950
BA = np.array([14692, 12896]) - 50
YD = np.array([12896, 11703]) - 50

axs[0].text(x=np.mean(HS1), y=0.99 * np.max(axs[0].get_ylim()), s='HS1', ha='center', va='top',rotation=90)
axs[0].text(x=np.mean(BA), y=0.99 * np.max(axs[0].get_ylim()), s='BA', ha='center', va='top',rotation=90)
axs[0].text(x=np.mean(YD), y=0.99 * np.max(axs[0].get_ylim()), s='YD', ha='center', va='top',rotation=90)

# Define the grey color with a transparency using RGBA tuple
grey_color = (0.8, 0.8, 0.8, 0.5)  # Grey color with 50% transparency.

# Function to add grey bars to the subplots
def add_grey_bars(ax):

    ax.axvspan(HS1[1], HS1[0], color='darkslategray', alpha=0.2, lw=0,zorder=-1)
    ax.axvspan(BA[1], BA[0], color='pink', alpha=0.2, lw=0,zorder=-1)
    ax.axvspan(YD[1], YD[0], color='navy', alpha=0.2, lw=0,zorder=-1)

for ax in axs:
    add_grey_bars(ax)


top = axs[0].get_position().y1
bottom = axs[-1].get_position().y0
left = axs[0].get_position().x0  # Assuming the left y-axis is used
right = axs[0].get_position().x1  # Assuming the right y-axis is not extended

box = patches.FancyBboxPatch(
    (left, bottom), right-left, top-bottom,
    transform=fig.transFigure, boxstyle="square,pad=0", 
    ec="black", fc="none", lw=1.8, clip_on=False
)
# Add the box to the figure
fig.patches.extend([box])


# Add legend to the last subplot
# axs[-1].legend(loc='upper left')

plt.show()
```

```python
# Refining the provided code to add the error range of the mean curve of the weighted average curves
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


# Convert the dictionary values to a list of arrays
sic_curves = list(weighted_avg_sic.values())

# Check that all arrays are of the same length
if all(len(curve) == len(sic_curves[0]) for curve in sic_curves):
    # Calculate the mean curve and its standard deviation (as error range)
    mean_sic_curve = np.mean(sic_curves, axis=0)
    std_sic_curve = np.std(sic_curves, axis=0)
else:
    raise ValueError("The SIC curves are not of the same length.")

plt.figure(figsize=(8, 5), dpi=600)

# Plot the mean curve
plt.plot(ds_seaice['age'], mean_sic_curve, color='darkgreen', label='LGMR SIC (fraction)', marker='o')

# Add the error range as shaded area
plt.fill_between(ds_seaice['age'], 
                 mean_sic_curve - std_sic_curve, 
                 mean_sic_curve + std_sic_curve, 
                 color='darkgreen', alpha=0.2)

# Add x and y labels
plt.xlabel('Age (yr BP)')
plt.ylabel('LGMR SIC (fraction)',color='darkgreen')

# Add a twin x-axis to plot the PC1
ax2 = plt.gca().twinx()
ax2.plot(df_pc['Age (ka)']*1000, df_pc['PC 1'], color='m', marker='^')
ax2.set_ylabel('Proxy data PC 1', color='m')

# Set the x-axis limits
plt.xlim(11000, 30000)

# Invert the x-axis
plt.gca().invert_xaxis()

# Add a box to the figure with a line width of 2
for _, spine in plt.gca().spines.items():
    spine.set_linewidth(2)

plt.show()

```
