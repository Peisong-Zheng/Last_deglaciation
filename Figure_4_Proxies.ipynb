# Load composite Pa/Th data for AMOC strength
Ref: 
1. Bohm, E. et al. Strong and deep Atlantic meridional overturning circulation during the last glacial cycle. Nature https://doi.org/10.1038/Nature14059 (2015)
2. McManus, J. F., Francois, R., Gherardi, J. M., Keigwin, L. D. & Brown-Leger, S. Collapse and rapid resumption of Atlantic meridional circulation linked to deglacial climate changes. Nature 428, 834–837 (2004).
3.  Henry, L. G. et al. North Atlantic ocean circulation and abrupt climate change during the last glaciation. Science 353, 470–474 (2016).
4.  Yu, J., Anderson, R.F., Jin, Z.D. et al. Millennial atmospheric CO2 changes linked to ocean ventilation modes over past 150,000 years. Nat. Geosci. (2023). https://doi.org/10.1038/s41561-023-01297-x

```python
# read AMOC data from 'Yu et al-2023-AMOC data.xlsx' under data_fig4 folder

import pandas as pd

file_path = 'data_fig4/Yu et al-2023-AMOC data.xlsx'

# Read the Excel file
df_AMOC = pd.read_excel(file_path,skiprows=3)
# rename the columns to 'age','age_scaling','combined_age','Pa_Th','error','refs'
df_AMOC.columns = ['age','age_scaling','combined_age','Pa_Th','error','refs']

# extract the 'combined_age','Pa_Th','error' columns
df_AMOC = df_AMOC[['combined_age','Pa_Th','error']]
# combined_age*1000 to convert to years
df_AMOC['combined_age'] = df_AMOC['combined_age']*1000

# save the data to csv file
df_AMOC.to_csv('data_fig4/AMOC.csv',index=False)
df_AMOC
```

# Load CO2 data
Ref:

2. Marcott et al., 2014, DOI: 10.1038/nature13799


```python
# Read CO2 data
import pandas as pd

file_path = 'data_fig4/Bauska et al-2021-Nature-DATA-CO2.xlsx'

# Read the Excel file
df_co2 = pd.read_excel(file_path,sheet_name='Sheet1')
# rename columns to 'age','co2','error'
df_co2.columns = ['age','co2','error']
# save the data to csv file
df_co2.to_csv('data_fig4/co2.csv',index=False)
df_co2
```

# Load CH4 data
Ref:

```python
# read the CH4 data Rhodes et al-2015-wais2015ch4.txt from data_fig4 folder
import pandas as pd

file_path = 'data_fig4/Rhodes et al-2015-wais2015ch4.txt'

df_ch4 = pd.read_csv(file_path,skiprows=98,sep='\t')
# rename columns to 'depth','age','age_kyr','ch4'
df_ch4.columns = ['depth','age','age_kyr','ch4']
# extract the 'age','ch4' columns
df_ch4 = df_ch4[['age','ch4']]
# save the data to csv file
df_ch4.to_csv('data_fig4/ch4.csv',index=False)
df_ch4
```

# Load the processed data

```python
import pandas as pd
df_AMOC = pd.read_csv('data_fig4/AMOC.csv')
df_co2 = pd.read_csv('data_fig4/co2.csv')
df_ch4 = pd.read_csv('data_fig4/ch4.csv')
df_co3 = pd.read_csv('data_fig4/co3.csv')
df_opal = pd.read_csv('data_fig4/opal.csv')
```

# Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Periods for the vertical grey bars
HS1 = np.array([17480, 14692]) - 50  # convert to b1950
BA = np.array([14692, 12896]) - 50
YD = np.array([12896, 11703]) - 50


# Function to add grey bars to the subplots
def add_grey_bars(ax):

    ax.axvspan(HS1[1], HS1[0], color='darkslategray', alpha=0.2, lw=0,zorder=-1)
    ax.axvspan(BA[1], BA[0], color='pink', alpha=0.2, lw=0,zorder=-1)
    ax.axvspan(YD[1], YD[0], color='navy', alpha=0.2, lw=0,zorder=-1)

# Plotting
fig, axs = plt.subplots(4, 1, sharex=True, figsize=(5,9),dpi=600)

# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

# Plot AMOC - Change marker to open circle
axs[0].errorbar(df_AMOC['combined_age'], df_AMOC['Pa_Th'], yerr=df_AMOC['error'], fmt='o', mfc='none')
axs[0].plot(df_AMOC['combined_age'], df_AMOC['Pa_Th'], '-', color='black')
axs[0].set_ylabel('$^{231}$Pa/$^{230}$Th')
axs[0].tick_params(labelbottom=False)
axs[0].spines['bottom'].set_visible(False)
axs[0].grid(True, linestyle='--')
axs[0].set_ylim(0.05,0.1)
axs[0].invert_yaxis()

axs[0].text(x=np.mean(HS1), y=0.055, s='HS1', ha='center', va='top')
axs[0].text(x=np.mean(BA), y=0.055, s='BA', ha='center', va='top')
axs[0].text(x=np.mean(YD), y=0.055, s='YD', ha='center', va='top')

# Plot CO2 - Original data
axs[1].errorbar(df_co2['age'], df_co2['co2'], yerr=df_co2['error'], color='green', alpha=0.3)

# Interpolate CO2 data to 10 year resolution
ages_interp = np.arange(df_co2['age'].min(), df_co2['age'].max(), 10)
co2_interp = np.interp(ages_interp, df_co2['age'], df_co2['co2'])

# Smooth the data using a 200-year moving average
window = 200 // 10  # Number of points in the window
co2_smoothed = np.convolve(co2_interp, np.ones(window)/window, 'same')

# Plot the smoothed CO2 data
axs[1].plot(ages_interp, co2_smoothed, color='green')

axs[1].set_ylabel('CO$_2$ (ppm)')
axs[1].tick_params(labelbottom=False)
axs[1].spines['bottom'].set_visible(False)
axs[1].spines['top'].set_visible(False)
axs[1].grid(True, linestyle='--')
axs[1].yaxis.tick_right()
axs[1].yaxis.set_label_position("right")

# Plot CO3 - Change marker and add legend
axs[2].plot(df_co3['age'], df_co3['co3'], '-', color='orange', label='$CO_3$$^2$$^-$', marker='s')
axs[2].set_ylabel('$CO_3$$^2$$^-$(μmol/kg)')
axs[2].tick_params(labelbottom=False)
axs[2].spines['bottom'].set_visible(False)
axs[2].spines['top'].set_visible(False)
axs[2].grid(True, linestyle='--')
axs[2].legend(loc='upper left')

# Plot Opal Flux - Change marker and add legend
axs2 = axs[2].twinx()
axs2.plot(df_opal['age'], df_opal['opal_flux'], '-', color='blue', label='Opal Flux', marker='^')
axs2.set_ylabel('Opal Flux (g/$cm^2$/kyr)')
axs2.spines['bottom'].set_visible(False)
axs2.spines['top'].set_visible(False)
axs2.legend(loc=[0.01,0.6])

# Plot CH4
axs[3].plot(df_ch4['age'], df_ch4['ch4'], '-', color='red')
axs[3].set_ylabel('$CH_4$ (ppb)')
axs[3].set_xlabel('Age (yr BP)')
axs[3].spines['top'].set_visible(False)
axs[3].grid(True, linestyle='--')
axs[3].set_xlim(11000,25000)
axs[3].invert_xaxis()
axs[3].yaxis.tick_right()
axs[3].yaxis.set_label_position("right")


# Label for subplots
axs[0].text(0.1, 0.97, 'a', transform=axs[0].transAxes, fontsize=12, fontweight='bold', va='top', ha='right')
axs[1].text(0.1, 0.7, 'b', transform=axs[1].transAxes, fontsize=12, fontweight='bold', va='top', ha='right')
axs[2].text(0.1, 0.45, 'c', transform=axs[2].transAxes, fontsize=12, fontweight='bold', va='top', ha='right')
axs[3].text(0.1, 0.5, 'd', transform=axs[3].transAxes, fontsize=12, fontweight='bold', va='top', ha='right')

# Add fancy box
top = axs[0].get_position().y1
bottom = axs[-1].get_position().y0
left = axs[0].get_position().x0
right = axs[0].get_position().x1

box = patches.FancyBboxPatch(
    (left, bottom), right-left, top-bottom,
    transform=fig.transFigure, boxstyle="square,pad=0",
    ec="black", fc="none", lw=1.6, clip_on=False
)
fig.patches.extend([box])

for ax in axs:
    add_grey_bars(ax)

# Show plot
plt.show()
```
