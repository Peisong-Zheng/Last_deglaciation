```python
import numpy as np
import matplotlib.pyplot as plt
from PyEMD import EMD

# Create a sample signal: a sum of two sine waves with different frequencies
time = np.linspace(0, 1, 400)
signal = 3*np.sin(2 * np.pi * 5 * time) + 2*np.sin(2 * np.pi * 13 * time)

# Execute EMD on the signal
emd = EMD()
IMFs = emd.emd(signal, time)

# Plotting
plt.figure(figsize=(10, 8))
plt.subplot(len(IMFs) + 1, 1, 1)
plt.plot(time, signal, 'r')
plt.title("Original Signal")

for n, imf in enumerate(IMFs, start=1):
    plt.subplot(len(IMFs) + 1, 1, n+1)
    plt.plot(time, imf, 'g')
    plt.title("IMF {}".format(n))

plt.tight_layout()
plt.show()
```
